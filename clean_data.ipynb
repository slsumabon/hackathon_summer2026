```python
import pandas as pd
```

```python
data = pd.read_csv("copy-of-2025-enhanced-statewide-summary.csv")
```

```python
data.head()
```

```python
dallas_all = data[(data["County"] == "DALLAS") & (data["Charter"] == "No")]
```

```python
dallas_some = dallas_all[['District',
 'Campus',
 'Region',
 'County',
 'School\nType',
 'Grades\nServed',
 'Number of\nStudents',
 '%\nEconomically\nDisadvantaged',
 '% EB/EL\nStudents',
 'Overall\nScore',
 'Student\nAchievement\nScore',
 'School\nProgress\nScore',
 'Academic\nGrowth\nScore',
 'Relative\nPerformance\nScore',
 'Closing\nthe Gaps Score',
 'Public\nEducation\nGrant',
 'Campus Enrollment\nType']]
```

```python
dallas_some.head()
```

```python
dallas_some["Public\nEducation\nGrant"].unique()
```

```python
dallas_hs = dallas_some[dallas_some["Grades\nServed"].str.contains("12", case=False, na=False)]
```

```python
dallas_hs = dallas_hs.rename(columns={"School\nType": "Type", 
                                      "Number of\nStudents": "Students",
                                     "Grades\nServed": "grades_served",
                                     "%\nEconomically\nDisadvantaged": "perc_ed",
                                     "% EB/EL\nStudents": "perc_eb_el",
                                     "Overall\nScore": "overall_score",
                                     "Student\nAchievement\nScore": "student_ach_score",
                                     "School\nProgress\nScore": "school_prog_score",
                                     "Academic\nGrowth\nScore": "academic_growth_score",
                                     "Relative\nPerformance\nScore": "relative_perf_score",
                                     "Closing\nthe Gaps Score": "closing_gap_score",
                                     "Public\nEducation\nGrant": "Grant",
                                     "Campus Enrollment\nType": "enroll_type"})
```

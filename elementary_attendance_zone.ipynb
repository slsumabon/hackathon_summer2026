```python
import geopandas as gpd
import folium

gdf = gpd.read_file("Elementary_Attendance_Boundaries.geojson")

center = gdf.geometry.union_all().centroid
m = folium.Map(location=[center.y, center.x], zoom_start=11)

folium.GeoJson(
    gdf,
    tooltip=folium.GeoJsonTooltip(fields=["ELEM_DESC"], aliases=["School:"]),
    style_function=lambda x: {
        "fillColor": "#3186cc",
        "color": "black",
        "weight": 1,
        "fillOpacity": 0.3
    }
).add_to(m)

m.save("dallas_school_zones.html")
```
